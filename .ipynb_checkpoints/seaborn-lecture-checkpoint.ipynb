#### Types of data:

- Continuous Data:- Numeric data with possibly infinite resolution
    - Height
    - Weight
    - total_bill and tip
    - temperature
- Categorical - Distinct categories
    - weekdays
    - Gender
    - 'smoker'
    - letter grades (A, B, C..)

- Discrete - Distinct numeric categories 
     - party_size - numeric but discete
     - number of customer complaints
     - number of flaws or defects.

### Relational Plots 

```python
# Explore relationship between total_bill and tip

sns.relplot(data = tips, x = 'total_bill', y = 'tip')

# returns a FacetGrid (which means a figure level object)
```

```python
# Set style
sns.set()

sns.relplot(data = tips, x = 'total_bill', y = 'tip')

plt.title("Total Bill vs Tip Amount")
```

```python
# Correlation with smokers
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue='smoker')
```

```python
# Correlation of time of day
# hue probably only effective with 2 or max of 3
# style changes shape of type of 
sns.relplot(data = tips, x = 'total_bill', y = 'tip', style='time')
```

```python
# col=  creates subplots of the data
# col_wrap= to specify # of columns
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue='time', col='day')
```

#### Key takeaways?
- Tip amoumt generallt increases with total_bill
- Works lunch on Thursday and Dinner on weekend

#### Ways to add DF columns to the chart
- hue
- col - creates subplots
- style


#### Regression plots

```python
# Add Regression line to relplot
# sns.lmplot = 
sns.lmplot(x = 'total_bill', y = 'tip', hue='smoker', data = tips)

# The opaque area if the 95% confidence interval
```

### Categorial Plots

#### Boxplots

```python
sns.boxplot(data = tips, x = 'tip')


plt.grid(True, ls = '--')
plt.xticks(np.arange(0, 10, 1));

# Line in box = median
# box = interquartile range (IQR)
# Outside lines "whiskers" = 'Min (Q1 - 1.5*IQR), Max(Q3 + 1.5*IQR)
# Dots outside whiskers are OUTLIERS
```

```python
tips.tip.describe()
```

```python
plt.figure(figsize=(10,8))
plt.subplot(221)
sns.boxplot(x='tip', data = tips)

plt.subplot(222)
sns.boxplot(y='tip', data = tips)

plt.subplot(223)
sns.boxplot(y='tip', x = 'size', data = tips)

plt.subplot(224)
sns.boxplot(x = 'tip', data = tips, y = 'time', hue = 'smoker')
```

#### Heatmap

```python
data = pd.crosstab(tips.time, tips['size'])
data
```

```python
sns.heatmap(data, cmap = plt.cm.Greens, annot=True, fmt = 'd')
```

```python
data = pd.crosstab(tips.time, tips.smoker)
data
```

```python
sns.heatmap(data, annot=True, fmt = 'd')
```

#### Pairplot

```python
sns.pairplot(tips)

# GREAT for looking for trends
```

```python
sns.pairplot(tips, hue='smoker')
```

#### Jointpplot

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip')

# scatter plots + the horizontal and vertical distributions
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind='reg')

# add regression line 
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind='hex')
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'kde')
```
