4. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
from env import host, user, password

chipotle_url = f'mysql+pymysql://{user}:{password}@{host}/chipotle'
    
query_chip = '''
SELECT * 
FROM orders
'''  

chipotle = pd.read_sql(query_chip, chipotle_url)
```

```python
chipotle.head()
```

create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
chipotle['item_price'] = chipotle['item_price'].str.replace('$','').astype(float)
```

```python
most_popular_and_revenue = chipotle.groupby('item_name').agg({'quantity': 'count', 'item_price': 'sum'}).sort_values(by='quantity', ascending=False).head()
```

```python
most_popular_and_revenue
```

```python
labels = ['Chicken Bowl', 'Chicken Burrito', 'Chips and Guac', 'Steak Burrito']

revenue = [7342.73, 5575.82, 2201.04, 3851.43]

plt.bar(labels, revenue, color='teal')
plt.xticks(rotation = -45)

plt.show()
```

```python
best_sellers = chipotle.groupby('item_name').quantity.sum().nlargest(4, keep='all')
best_sellers.index
```
