1. Use matplotlib to plot the following equation:

```y = x^2 âˆ’ x + 2```

- You'll need to write the code that generates the x and y points.

- Add an anotation for the point 0, 0, the orig
in.

````python
x = list(range(-10, 10))
y = [x ** 2 - x + 2 for x in x]

plt.scatter(x, y)
plt.plot(x, y, ls = "--")

plt.text(0, 0, '(0,0)', fontsize=10, color='blue')

plt.show()
````

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

    - ```y = âˆšx```
    - ```y = x^3```
    - ```y = 2^x```
    - ```y = 1 / (x + 1)```
- You can use functions from the ```math``` module to help implement some of the equations above.


````python
x = list(range(0,10))
y1 = [x ** (1/2) for x in x]

y2 = [x ** 3 for x in x]

y3 = [2 ** x for x in x]

y4 = [1 / (x + 1) for x in x]

plt.plot(x, y1)
plt.show()

plt.plot(x, y2)
plt.show()

plt.plot(x, y3)
plt.show()

plt.plot(x, y4)
plt.show()
````

3. Combine the figures you created in the last step into one large figure with 4 subplots.

````python
plt.subplot(1, 4, 1)
plt.plot(x, y1)
plt.title('y1')

plt.subplot(1, 4, 2)
plt.plot(x, y2)
plt.title('y2')

plt.subplot(1, 4, 3)
plt.plot(x, y3)
plt.title('y3')

plt.subplot(1, 4, 4)
plt.plot(x, y4)
plt.title('y4')

plt.subplots_adjust(left=None, bottom=None, right=2, top=None, wspace=.25, hspace=1)


plt.show()
````

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

````python
plt.subplot(2, 2, 1)
plt.plot(x, y1, c = 'blue')
plt.title('y1')

plt.subplot(2, 2, 2)
plt.plot(x, y2, c = 'orange')
plt.title('y2')

plt.subplot(2, 2, 3)
plt.plot(x, y3, c = 'red')
plt.title('y3')

plt.subplot(2, 2, 4)
plt.plot(x, y4, c = 'green')
plt.title('y4')

plt.suptitle('Matplotlib subplot exercise')
plt.subplots_adjust(left=None, bottom=None, right=1, top=None, wspace=.25, hspace=.5)



plt.show()
````
