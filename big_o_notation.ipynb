```python


x = list(range(1, 15))  
y1 = [0 * n + 1 for n in x]
y2 = [math.log(n, 10) for n in x]
y3 = [n for n in x]
y4 = [n * math.log(n, 10) for n in x]
y5 = [n ** 2 for n in x]
y6 = [2 ** n for n in x]
y7 = [math.factorial(n) for n in x]
y8 = [n ** n for n in x]


# 'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'
plt.figure(figsize=(10, 6))



plt.plot(x, y1, label='$\mathcal{O}(1)$', c = 'b')

plt.plot(x, y2, label='$\mathcal{O}(log n)$', c = 'g')

plt.plot(x, y3, label='$\mathcal{O}(n)$', c = 'r')

plt.plot(x, y4, label='$\mathcal{O}(n log n)$', c = 'c')

plt.plot(x, y5, label='$\mathcal{O}(n^2)$', c = 'm')

plt.plot(x, y6, label='$\mathcal{O}(2^n)$', c = 'y')

plt.plot(x, y7, label = '$\mathcal{O}(n!)$', c = 'k')

plt.plot(x, y8, label = '$\mathcal{O}(n^n)$', c = 'salmon')

plt.legend(loc = 'upper right')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.title('Big O Notation')

plt.xlim(1,11)
plt.ylim(0,15)

plt.show()
```

```python
 # SUBPLOTS 

plt.subplot(241)
plt.plot(x, y1, label = '$O(1)$')
plt.title('O(1)')

plt.subplot(242)
plt.plot(x, y2)
plt.title('O(log n)')

plt.subplot(243)
plt.plot(x, y3)
plt.title('O(n)')

plt.subplot(244)
plt.plot(x, y4)
plt.title('O(n log n)')

plt.subplot(245)
plt.plot(x, y5)
plt.title('O(n^2)')

plt.subplot(246)
plt.plot(x, y6)
plt.title('O(2^n)')

plt.subplot(247)
plt.plot(x, y7)
plt.title("O(n!)")

plt.subplot(248)
plt.plot(x, y8)
plt.title('O(n^n)')

plt.subplots_adjust(left=None, bottom=None, right=1.5, top=1, wspace=1, hspace=1)
# plt.xticks()


plt.show()
```
